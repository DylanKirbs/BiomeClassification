# Exploratory Data Analysis

```python
# Import the data constants
from constants import RESOLUTIONS, VARIABLES
from itertools import product as cartesianProduct
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from dataUtils import readGeoData, getFilePath, plotHeatmap
```

```python
# Data location
dataDir = "./data"
variables = ['average temperature', 'precipitation']
resolutions = ['5 minutes']

fileList = [(VARIABLES[variable], RESOLUTIONS[res]) for variable, res in cartesianProduct(variables, resolutions)]
```

```python
# Get the Ave Temp data for January
filePath_temperature = getFilePath("./", fileList[0][0], fileList[0][1], '01')
meta_temperature, data_temperature = readGeoData(filePath_temperature)

# Get the Precipitation data for January
filePath_precipitation = getFilePath("./", fileList[1][0], fileList[1][1], '01')
meta_precipitation, data_precipitation = readGeoData(filePath_precipitation)
```

```python
plotHeatmap(plt, sns, data_temperature, 'Average Temperature')
plt.show()
```

```python
plotHeatmap(plt, sns, data_precipitation, 'Precipitation')
```
